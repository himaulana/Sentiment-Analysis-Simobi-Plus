## Scrapping Reviews Simobi from Play Store

```python
%pip install google-play-scraper
%pip install pandas
```

```python
import pandas as pd

from google_play_scraper import Sort, reviews_all

reviews = reviews_all(
    "com.simas.mobile.SimobiPlus",
    lang="id",
    country="id",
    sort=Sort.NEWEST,
    sleep_milliseconds=100,
)

reviews_df = pd.DataFrame(reviews)
reviews_df.shape
```

```python
reviews_df.info()
```

```python
reviews_df.head()
```

```python
reviews_df.isna().sum()
```

```python
df = reviews_df.dropna(subset=["content"]).drop_duplicates(subset=["content"])
df = df[["content"]].copy()
df.info()
```

```python
df.to_csv("../data/reviews_simobi.csv", index=False, encoding="utf-8-sig")
```
